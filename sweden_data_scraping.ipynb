```python
#Trying to find the table that I want
tables[2]
```

```python
sweden_1 = tables[2]
sweden_1.columns[2]
```

```python
#Renaming some of the columns
sweden_1 = sweden_1.rename(columns={'Song[2]': 'Song' , 'Points.1': 'Points SF'})
sweden_1
```

```python
#Dropping the last row (index 63) since we can't know the results from 2024
sweden_1 = sweden_1.drop(63)
sweden_1
```

```python
#Saving my tables into a CSV file
sweden_1.to_csv('sweden_data.csv', index=True) #index= true in case I need the index column 
```
